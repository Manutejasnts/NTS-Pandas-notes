# 8)Merge in DataFrame

```python
import pandas as pd

df1 = pd.DataFrame({
    "City":["new york","chicago","orlando"],
    "temperature":[21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["chicago","new york","orlando"],
    "Humidity":[65,68,75]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="City")
df3
```

# If the data are not matching

```python
import pandas as pd

df1 = pd.DataFrame({
    "City":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    "City":["chicago","new york","san francisco"],
    "Humidity":[65,68,71]
})
df2
```

```python
df = pd.merge(df1,df2,on="City")
df
```

```python
df = pd.merge(df1,df2,on="City",how="outer")
df
```

```python
df = pd.merge(df1,df2,on="City",how="inner")
df
```

```python
df = pd.merge(df1,df2,on="City",how="left")
df
```

```python
df = pd.merge(df1,df2,on="City",how="right")
df
```

```python
df = pd.merge(df1,df2,on="City",how="outer",indicator=True)
df
```

# Another example of Merge

```python
import pandas as pd

df1 = pd.DataFrame({
    "City":["chicago","new york","san diego"],
    "temperature":[21,14,35],
    "Humidity":[65,68,71]
})
df1
```

```python
import pandas as pd

df2 = pd.DataFrame({
    "City":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38],
    "Humidity":[65,68,71,75]
})
df2
```

```python
df = pd.merge(df1,df2,on="City",how="outer",indicator=True)
df
```

```python
df = pd.merge(df1,df2,on="City",how="outer",indicator=True,suffixes=('_left','_right'))
df
```
