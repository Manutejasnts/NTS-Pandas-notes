```python
import pandas as pd
df = pd.read_csv("Supporting files//Handling missing data.csv",parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'events':'No event'})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

```python
new_df = df.fillna(method='pad')
new_df
```

```python
new_df = df.fillna(method='bfill',axis='columns')
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df = df.fillna(method="ffill",limit=2)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = df.interpolate(method='linear')
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how = "all")
new_df
```
