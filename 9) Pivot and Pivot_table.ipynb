# PIVOT

```python
import pandas as pd
df = pd.read_csv("Supporting files//pivot table.csv")
df
```

```python
df.pivot(index="data",columns="city")
```

```python
df.pivot(index="data",columns="city",values="humidity")
```

```python
df.pivot(index="humidity",columns="city")
```

```python
df = pd.DataFrame({
    "date":["5/1/2017","5/1/2017","5/2/2017","5/2/2017","5/1/2017","5/1/2017","5/2/2017","5/2/2017"],
    "city":["new york","new york","new york","new york","mumbai","mumbai","mumbai","mumbai"],
    "temperature":[65,61,70,72,75,78,82,80],
    "humidity":[56,54,60,62,80,83,85,26]
})
df
```

```python
df.pivot_table(index="city",columns="date")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="sum")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="count")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="diff")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="mean")
```

```python
df.pivot_table(index="city",columns="date",aggfunc="sum",margins=True)
```

```python
df.pivot_table(index="city",columns="date",aggfunc="mean",margins=True,margins_name="Total")
```
