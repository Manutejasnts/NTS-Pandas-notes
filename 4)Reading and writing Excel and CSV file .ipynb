# How to use converters in pandas

```python
import pandas as pd
df =  pd.read_csv('C:\\Users\\NATARAJ\\Python Libaries NTS Notes\\NTS Pandas Notes\\Supporting files\\Reading_Writing_Excel_csv_files.csv')
df
```

```python
import pandas as pd

def convert_people_cell(cell):
    if cell=='n.a.':
        return 'Tejas-N'
    return cell

def convert_eps_cell(cell):
    if cell=='not available':
        return None
    return cell
def convert_ticket_cell(cell):
    if cell=='WMT':
        return 'NTS'
    return cell

df = pd.read_csv('C:\\Users\\NATARAJ\\Python Libaries NTS Notes\\NTS Pandas Notes\\Supporting files\\Reading_Writing_Excel_csv_files.csv',converters={
    'people':convert_people_cell,
    'eps':convert_eps_cell,
    'tickets':convert_ticket_cell
})
df
```

```python
import pandas as pd
df_stocks = pd.DataFrame({
     'tickets':['GOOGL','WMT','MSFT'],
     'price':[845,65,64],
     'pe':[30.37,14.26,30.97],
     'eps':[27.82,4.61,2.12]
})
df_weather = pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[32,35,28],
    'event':['Rain','Sunny','Snow']
})

```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name='stocks')
    df_weather.to_excel(writer,sheet_name='weather')
```
